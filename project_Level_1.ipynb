```python
import pandas as pd
```

```python
data_full = pd.read_csv('/content/gdrive/MyDrive/House_price/Housing.csv')
```

```python
data_full.head(545)
```

```python
data_full["guestroom"] = data_full["guestroom"].replace({"yes": 1, "no": 0})
data_full["mainroad"] = data_full["mainroad"].replace({"yes": 1, "no": 0})
data_full["basement"] = data_full["basement"].replace({"yes": 1, "no": 0})
data_full["hotwaterheating"] = data_full["hotwaterheating"].replace({"yes": 1, "no": 0})
data_full["airconditioning"] = data_full["airconditioning"].replace({"yes": 1, "no": 0})
data_full["prefarea"] = data_full["prefarea"].replace({"yes": 1, "no": 0})
data_full["furnishingstatus"] = data_full["furnishingstatus"].replace({"furnished": 1, "semi-furnished": 0.8, "unfurnished":0})
```

```python
data_full
```

```python

from sklearn.utils import shuffle

shu_data_full = shuffle(data_full, random_state=42)
data_full
```

```python
data=  shu_data_full.iloc[:, 1:13]
target=  shu_data_full.iloc[:, [0]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, Y_pred)
```

```python
mse
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, Y_pred)
r2
```
