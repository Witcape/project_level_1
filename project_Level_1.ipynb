```python
import pandas as pd
```

```python
data_full = pd.read_csv('/content/gdrive/MyDrive/House_price/Housing.csv')
```

```python
data_full.head(545)
```

```python
data_full["guestroom"] = data_full["guestroom"].replace({"yes": 1, "no": 0})
data_full["mainroad"] = data_full["mainroad"].replace({"yes": 1, "no": 0})
data_full["basement"] = data_full["basement"].replace({"yes": 1, "no": 0})
data_full["hotwaterheating"] = data_full["hotwaterheating"].replace({"yes": 1, "no": 0})
data_full["airconditioning"] = data_full["airconditioning"].replace({"yes": 1, "no": 0})
data_full["prefarea"] = data_full["prefarea"].replace({"yes": 1, "no": 0})
data_full["furnishingstatus"] = data_full["furnishingstatus"].replace({"furnished": 1, "semi-furnished": 0.8, "unfurnished":0})
```

```python
data_full
```

```python

from sklearn.utils import shuffle

shu_data_full = shuffle(data_full, random_state=42)
data_full
```

```python
data=  shu_data_full.iloc[:, 1:13]
target=  shu_data_full.iloc[:, [0]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data_scaler = StandardScaler()
target_scaler = MinMaxScaler()

scaled_data = data_scaler.fit_transform(data)
scaled_target = target_scaler.fit_transform(target)

data = scaled_data
target = scaled_target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=50)
```

```python
X_train.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear,relu,sigmoid
```

```python
model = Sequential(
    [
        tf.keras.Input(shape = (12,)),
        Dense(25, activation = 'relu'),
        Dense(10, activation = 'relu'),
        Dense(1, activation = 'linear')
    ]
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(),
    loss = 'mean_squared_error',
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs = 40
)
```

```python
loss = history.history['loss']
epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
